# Task
Show an example of training a simple ML model.

## Generate data

### Subtask:
Create some synthetic data for a simple regression problem.


**Reasoning**:
The first step is to import the necessary library, which is numpy, and then generate the synthetic data for X and y as described in the instructions.



```python
import numpy as np

# Generate 100 data points for X
X = np.random.rand(100, 1) * 10

# Generate the corresponding target variable y with noise
noise = np.random.randn(100, 1)
y = 2 * X + 1 + noise

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

## Define model

### Subtask:
Define a simple linear regression model using a machine learning library.


**Reasoning**:
Import the LinearRegression class and instantiate it.



```python
from sklearn.linear_model import LinearRegression

# Instantiate a LinearRegression object
model = LinearRegression()
```

## Train model

### Subtask:
Train the model on the generated data.


**Reasoning**:
Fit the linear regression model to the generated training data.



```python
model.fit(X, y)
print("Model trained successfully.")
```

## Evaluate model

### Subtask:
Evaluate the trained model's performance.


**Reasoning**:
Import the necessary function, make predictions, calculate the MSE, and print the result.



```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)

print(f"Mean Squared Error: {mse}")
```

## Summary:

### Data Analysis Key Findings

*   Synthetic data was generated for a simple regression problem, consisting of 100 data points for feature X and a corresponding target variable y with added noise. The relationship is approximately linear: \(y \approx 2 * X + 1\).
*   A `LinearRegression` model from the `sklearn` library was successfully defined and instantiated.
*   The defined `LinearRegression` model was trained using the generated synthetic data (`X` and `y`).
*   The trained model's performance was evaluated using the Mean Squared Error (MSE) metric on the training data, resulting in an MSE of approximately 0.8825.

### Insights or Next Steps

*   The MSE of 0.8825 indicates that the model has learned the underlying linear relationship with some error, which is expected given the added noise in the synthetic data.
*   The next steps could involve splitting the data into training and testing sets for a more robust evaluation of the model's generalization performance and potentially examining the model's coefficients to see how well they match the true relationship (\(y = 2X + 1\)).
